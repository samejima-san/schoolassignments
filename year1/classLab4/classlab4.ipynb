```python
class Cards:
    #suit color rank
    #suit max = 4
    #rank max = 12
    
    def __init__(self, suit, rank):
        rankdict = {1:'ace', 2:'two', 3:'three', 4:'four', 5:'five', 6:'six',7:'seven',8:'eight',9:'nine',10:'jack',11:'queen',12:'king'}
        suitdict = {1:'spades',2:'hearts', 3:'clubs',4:'diamonds'}
            
        self.suit = suitdict.get(suit)
        self.rank = rankdict.get(rank)
        self.cardgif = ""
        self.rankvalue = rank
    
    def reverse_dict(rankdict):
        return {value:key for key, value in rankdict.items()}
    
    def reverse_dict(suitdict):
        return {value:key for key, value in suitdict.items()}
    
    def cardgiffun(self, suit, rank):
        if suit == 4:
            tempsuit = "D"
        elif suit == 3:
            tempsuit = "C"
        elif suit == 2:
            tempsuit = "H"
        else:
            tempsuit = "S"
        
        if rank == 1:
            temprank = "A"
        elif rank == 11:
            temprank = "J"
        elif rank == 10:
            temprank = "T"
        elif rank == 12:
            temprank = "Q"
        elif rank == 13:
            temprank = "K"
        else:
            temprank = str(rank)
        self.cardgif = tempsuit+temprank+".gif"      
            
    #displays the card
    def display(self):
        print(str(self.rank) + " of " + str(self.suit))
    
    def displayonwin(self, x,y):
        card = Image(Point(x,y),"Cards/"+self.cardgif)
        card.draw(win)
        #time.sleep(1)
        #card.undraw()
    
    def undrawCard(self):
        card.undraw()
    
```
